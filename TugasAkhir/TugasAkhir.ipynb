##**Configuring the chefboost**

```python
id3 = {'algorithm': 'ID3'}
c45 = {'algorithm': 'C4.5'}
```

##**Accuracy**

```python
model = chef.fit(df.copy(), c45)
```

```python
#Testing it on the Data from 2016 to get the accuracy
for index, instance in df2.iterrows():
  prediction = chef.predict(model, instance)
  actual = instance['Decision']
  
  if actual == prediction:
    classified = True
  else:
    classified = False
    print("*", end="")

  print(index,'Actual: ',actual, ' - ', 'C4.5 Prediction: ', prediction)
```
